```python
import ikpy.chain
import numpy as np
import ikpy.utils.plot as plot_utils
```

```python
front_right_chain = ikpy.chain.Chain.from_urdf_file("../../../../SpotSim/URDF/spot-front-right.urdf")
front_left_chain = ikpy.chain.Chain.from_urdf_file("../../../../SpotSim/URDF/spot-front-left.urdf")
rear_right_chain = ikpy.chain.Chain.from_urdf_file("../../../../SpotSim/URDF/spot-rear-right.urdf")
rear_left_chain = ikpy.chain.Chain.from_urdf_file("../../../../SpotSim/URDF/spot-rear-left.urdf")
```

```python
from mpl_toolkits.mplot3d import Axes3D
from ikpy.utils import plot
fig, ax = plot.init_3d_figure()
front_right_chain.plot([0] * (len(front_right_chain)), ax)
front_left_chain.plot([0] * (len(front_left_chain)), ax)
rear_right_chain.plot([0] * (len(rear_right_chain)), ax)
rear_left_chain.plot([0] * (len(rear_left_chain)), ax)
ax.legend()
```
