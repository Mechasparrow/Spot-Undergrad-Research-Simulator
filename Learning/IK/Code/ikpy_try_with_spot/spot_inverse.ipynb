```python
import ikpy.chain
import numpy as np
import ikpy.utils.plot as plot_utils
```

```python
front_right_chain = ikpy.chain.Chain.from_urdf_file("../../../../SpotSim/URDF/spot-front-left.urdf")
front_right_chain.active_links_mask[0] = False
print(front_right_chain)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from ikpy.utils import plot
fig, ax = plot.init_3d_figure()

print(front_right_chain.forward_kinematics([0,0,0,0])[:3,3])
front_right_chain.plot([0,0,0,0], ax)
ax.legend()
```
